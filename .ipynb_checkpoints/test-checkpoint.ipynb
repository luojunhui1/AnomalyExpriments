```python
from numpy import *
from tqdm import *
from sklearn.ensemble import IsolationForest

def get_centers(labels, flow):
    unique_labels = unique(labels)
    centers = []
    for label in unique_labels:
        centers.append(mean(flow[where(labels == label)], axis=0))
    return array(centers)

def cal_param(pred, real):
    count= 0.0
    for i in range(len(pred)):
        if pred[i] == 1 and pred[i] == real[i]:
            count = count + 1
    return count/len(where(pred == 1)[0]), count/len(where(real == 1)[0])

def iforest_ad(data, thresh = None):
    clf = IsolationForest(contamination=0.02, max_features=2)
    clf.fit(data)
    anomaly_score = clf.score_samples(data)
    anomaly_score = (anomaly_score-min(anomaly_score))/(max(anomaly_score)-min(anomaly_score))
    anomaly_score = 1 - anomaly_score
    if thresh == None:
        thresh = 90
    return where(anomaly_score >= percentile(anomaly_score, thresh))[0]

# get inflow, outflow data
inflow = load('./gendata/gen_inflows.npy')
outflow = load('./gendata/gen_outflows.npy')

inorm = norm_by_row(inflow)
onorm = norm_by_row(outflow)

aggl = load('./gendata/aggl.npy')
ad_real = load('./gendata/gen_anomalies.npy')
accs = []
recalls = []

tn = 6*7*24
rn = 256

for thresh in trange(79, 99):
    iforest_ads = array([])
    for t in range(0, tn):
        ad = array([0]*rn)
        for l in range(0, len(unique(aggl))):
            label_index = where(aggl == l)[0]
            io_norm = stack([inorm[label_index, t], onorm[label_index, t]], axis=1)
            ad_index = iforest_ad(io_norm, thresh)
            ad[label_index[ad_index]] = 1
        iforest_ads = concatenate([iforest_ads, ad])
    acc, recall = cal_param(iforest_ads, ad_real.T.flatten())
    accs.append(acc)
    recalls.append(recall)
```

```python
fig = plt.figure(figsize=(20, 20))
_, ax = plt.subplots()
ax.plot(range(0, 20), accs, c='r', ls=':', marker='s', label='acc')
ax.set_xlabel('k')
ax.legend()

ax1 = ax.twinx()
ax1.plot(range(0, 20), recalls, c='b', ls=':', marker='.', label='recall')
ax1.legend()

ax1.set_xticks(range(0, 20), tuple(range(79, 99)))

plt.show()
```

```python
from utils.BaseSVDD import *

def svdd_ad(data, thresh = None):
    svdd = BaseSVDD(C=thresh, gamma=0.3, kernel='rbf', display='off')
    svdd.fit(data)
    anomaly_score = svdd.get_distance(data) - svdd.radius
#     anomaly_score = (anomaly_score-min(anomaly_score))/(max(anomaly_score)-min(anomaly_score))
    anomaly_score = array(anomaly_score).squeeze()
    return where(anomaly_score > 0)[0]

# get inflow, outflow data
inflow = load('./gendata/gen_inflows.npy')
outflow = load('./gendata/gen_outflows.npy')

inorm = norm_by_row(inflow)
onorm = norm_by_row(outflow)

aggl = load('./gendata/aggl.npy')
ad_real = load('./gendata/gen_anomalies.npy')

accs = []
recalls = []

for thresh in trange(79, 99):
    iforest_ads = array([])
    for t in range(0, tn):
        ad = array([0]*rn)
        for l in range(0, len(unique(aggl))):
            label_index = where(aggl == l)[0]
            io_norm = stack([inorm[label_index, t], onorm[label_index, t]], axis=1)
            ad_index = svdd_ad(io_norm, thresh)
            ad[label_index[ad_index]] = 1
        iforest_ads = concatenate([iforest_ads, ad])
    acc, recall = cal_param(iforest_ads, ad_real.T.flatten())
    accs.append(acc)
    recalls.append(recall)
```

```python
def chisquare_ad(data, k = 3):
    means = mean(data,axis=1)
    chi = [sum(((data[i] - means)**2)/means) for i in range(0, len(data))]
    cur = chi - mean(chi)
    anomaly_score = [0 if cur[i] < 0 else cur[i] for i in range(0, len(cur))]
    return where(anomaly_score >= k)[0]

accs = []
recalls = []

for thresh in linspace(1, 3, 20,endpoint=False)::
    iforest_ads = array([])
    for t in range(0, tn):
        ad = array([0]*rn)
        for l in range(0, len(unique(aggl))):
            label_index = where(aggl == l)[0]
            io_norm = stack([inorm[label_index, t], onorm[label_index, t]], axis=1)
            ad_index = chisquare_ad(io_norm, thresh)
            ad[label_index[ad_index]] = 1
        iforest_ads = concatenate([iforest_ads, ad])
    acc, recall = cal_param(iforest_ads, ad_real.T.flatten())
    accs.append(acc)
    recalls.append(recall)
```

```python
def createData(data,datalen,col):
    xlist=[]
    dtlist=[]
    for i in range(data.index.min()+datalen-1, data.index.max()+1):
        xlist.append(data.loc[i-datalen+1:i].values)
        dtlist.append([i])
    xlist=np.array(xlist)
    dtlist=np.array(dtlist)
    print(xlist.shape,dtlist.shape)
    cols=['dt']+['%s_t-%d'%(col,r) for r in range(datalen,0,-1)]
    return pd.DataFrame(data=np.concatenate([dtlist,xlist],axis=1),columns=cols)
```

```python
cols=['inflow_{}'.format(i) for i in range(0, rn)] + ['outflow_{}'.format(i) for i in range(0, rn)] + ['is_weekday','is_holiday','week','time_order']
```

```python
def createSet(data,trcols,tscols, random_index):
    l=[]
    trs = np.array([])
    for area in range(0, 1024):
        l=[]
        for col in ["inflow", "outflow"]:
            fcols = ['%s_%d_t-%d'%(col, area, r) for r in range(timelong, ylong, -1)] 
            cur_ = data[fcols].values[random_index,:].reshape(len(random_index), xlong, 1) 
            cur_zi = cur_ == 0
            cur_[cur_zi] = 1 
            l.append(cur_)
        for col in ['is_weekday','is_holiday','week','time_order']:
            fcols = ['%s_t-%d'%(col, r) for r in range(timelong, ylong, -1)]   
            l.append(data[fcols].values[random_index,:].reshape(len(random_index), xlong, 1))
        tr = np.concatenate(l,axis=2)

        if trs.shape[0] == 0:
            trs = tr
        else:
            trs = np.concatenate([trs, tr],axis=0)

    l=[]
    tss = np.array([])
    for area in range(0, 1024):
        l=[]
        for col in ["inflow", "outflow"]:
            fcols = ['%s_%d_t-%d'%(col, area, r) for r in range(ylong, 0, -1)]
            cur_ = data[fcols].values[random_index,:].reshape(len(random_index), ylong, 1)
            cur_zi = cur_ == 0
            cur_[cur_zi] = 1 
            l.append(cur_)
        ts = np.concatenate(l,axis=2)
        
        if tss.shape[0] == 0:
            tss = ts
        else:
            tss = np.concatenate([tss, ts],axis=0)
    return trs,tss
```

```python
class seq2seq(nn.Module):
    def __init__(self, input_dim1, input_dim2, hid_dim, output_dim):
        super(seq2seq, self).__init__()
        self.embedding_hour = nn.Embedding(24,1)
        self.embedding_week = nn.Embedding(7,1)
        
        self.encode = nn.LSTM(input_dim1, hid_dim, num_layers = 1)
        self.decode = nn.LSTM(input_dim2, hid_dim, num_layers = 1)
        self.out = nn.Linear(hid_dim, output_dim)
    
    def forward(self, x, trg, isTrain):
        x1 = self.embedding_hour(x[:,:,-1].long())
        x2 = self.embedding_week(x[:,:,-2].long())
        encode_len = x.shape[0]
        batch_size = x.shape[1]
        
        x_input = torch.cat([x1, x2, x[:,:,:-2]], 2)
        
        outputs, encode_state = self.encode(x_input)
        
        decode_len = trg.shape[0]
        batch_size = trg.shape[1]
        outputs = torch.zeros(decode_len, batch_size, 2)
        
        for t in range(decode_len):
            if t==0:
                decode_input = torch.zeros_like(trg[0,:,:])
                h_next = encode_state
            else:
                decode_input = trg[t,:,:] if isTrain else prediction
            h_now = h_next
            decode_output, h_next = self.decode(decode_input.reshape(1, batch_size, 2), h_now)
            prediction = self.out(decode_output.squeeze(0))
            outputs[t] = prediction
        return outputs
```

```python
import torch.utils.data as Data

x_train = torch.from_numpy(X_train).float()
y_train = torch.from_numpy(y_train).float()
x_val = torch.from_numpy(X_val).float()
y_val = torch.from_numpy(y_val).float()
torch.manual_seed(42)
data_loader_train = Data.DataLoader(dataset = Data.TensorDataset(x_train, y_train), batch_size = batch_size, shuffle = True)
data_loader_val = Data.DataLoader(dataset = Data.TensorDataset(x_val, y_val), batch_size = batch_size, shuffle = True)
```

```python
model = seq2seq(x_train.shape[2], 2, 128, 2)
model=model.cuda()
epochs = 1000
lr = 0.001
optimizer = torch.optim.SGD(model.parameters(), lr = lr)
criterion = nn.MSELoss()
```

```python
def train_eval(flag, data_loader, model, optimizer, criterion):
    model.train() if flag == True else model.eval()
    predicts = []
    groud_turth = []
    loss_num = 0.0
    for i, (x, y) in enumerate(data_loader):
        x1 = x.permute(1, 0, 2)
        y1 = y.permute(1, 0, 2)
        x1, y1 = x1.cuda(), y1.cuda(),
        outputs = model(x1, y1, flag)
        outputs1 = outputs.permute(1, 0, 2)
        loss = criterion(outputs1, y)
        loss = loss.cpu()
        if flag == True:
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        groud_turth.extend(y.detach().numpy().tolist())
        predicts.extend(outputs1.detach().numpy().tolist())
        loss_num += loss.data
    loss_num = loss_num / len(data_loader)
    groud_turth = np.array(groud_turth).reshape(len(groud_turth)*len(groud_turth[0]), -1)
    predicts = np.array(predicts).reshape(len(predicts)*len(predicts[0]), -1)
    return loss_num, groud_turth, predicts
```

```python
min_mae = 0
min_mse = 1000000000
for epoch in range(epochs):
    train_loss,train_groud_turth, train_predicts = train_eval(True, data_loader_train, model, optimizer, criterion)
    test_loss, test_groud_turth, test_predicts= train_eval(False, data_loader_val, model, optimizer, criterion)
    print("epoch = %s, train: loss = %.4f, test: loss = %.4f"% (epoch, train_loss, test_loss))
    train_mae = mean_absolute_error(train_groud_turth, train_predicts)
    train_mse = mean_squared_error(train_groud_turth, train_predicts)
    test_mae = mean_absolute_error(test_groud_turth, test_predicts)
    test_mse = mean_squared_error(test_groud_turth, test_predicts)
    if test_mse < min_mse:
        min_mae = test_mae
        min_mse = test_mse
        min_epoch = epoch
        torch.save(model, 'model/ad.01_b128_lambda_10_gen.pkl')
    print("epoch = %s, train: loss = %.4f, mae = %.4f, mse = %.4f; test: loss = %.4f, mae = %.4f, mse = %.4f"% (epoch, train_loss, train_mae, train_mse, test_loss, test_mae, test_mse))
print("epoch = %s, min_mae = %.4f, min_mse = %.4f"% (min_epoch, min_mae, min_mse))
```
